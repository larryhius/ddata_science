```python
# Import Library
import matplotlib.pyplot as plt
```

```python
negara = ['Indonesia', 'Brunei Darussalam', 'Malaysia', 'Thailand', 'Vietnam']
jumlah_penduduk = [267, 5.63, 327, 126, 69.4]
```

```python
plt.figure(figsize=(10,6))
plt.plot(negara, jumlah_penduduk, marker='o', color='b', linestyle='-', linewidth=2, markersize=8)
plt.title('Jumlah Penduduk di Beberapa Negara Asia Tenggara', fontsize=14)
plt.xlabel('Negara', fontsize=12)
plt.ylabel('Jumlah Penduduk (juta)', fontsize=12)
plt.grid(True)
```
