```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
file_path = 'driver_income.json'  # Update this with your actual file path
data = pd.read_json(file_path)
```

```python
# Set style for visualizations
sns.set(style="whitegrid")

# Visualization 1: Income Distribution by Ownership Type
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='ownership', y='income')
plt.title("Income Distribution by Ownership Type")
plt.xlabel("Ownership Type")
plt.ylabel("Income")
```

```python
# Visualization 2: Average Distance vs. Average Income
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='avg_distance_normalized', y='income', hue='ownership')
plt.title("Average Distance vs. Income")
plt.xlabel("Average Distance (Normalized)")
plt.ylabel("Income")
plt.legend(title="Ownership Type")
```

```python
# Visualization 3: Tenure vs. Average Income
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='tenure', y='income', errorbar=None)
plt.title("Tenure vs. Average Income")
plt.xlabel("Tenure (Years)")
plt.ylabel("Average Income")
```
